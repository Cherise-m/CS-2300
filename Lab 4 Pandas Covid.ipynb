Find the parameter that has the highest correlation with 'deaths_covid'. The parameter is on_hand_supply_therapeutic_c_bamlanivimab_etesevimab_courses, which I think has something to do with covid deaths or affects them in some way. That's why it has the highest correalations with covid deaths.

```python
def crosscorr(df_x, df_y, lag= 0):
    """
    This is a lagging cross correalation function.
    lag: int 
    df_x, df_y: pandas.series objects of equal length
    return:crosscorr as a float
    """
    
    return df_x.corr(df_y.shift(lag))
```

```python
crosscorr(df_WI['staffed_adult_icu_bed_occupancy'], df_WI['previous_day_admission_adult_covid_confirmed'],8)
```

```python
corr_coefficient = []

for x in range(40):
    g = crosscorr(df_WI['staffed_adult_icu_bed_occupancy'], df_WI['previous_day_admission_adult_covid_confirmed'],x)
    corr_coefficient.append(g)
    
plt.plot(range(40),corr_coefficient)
```
